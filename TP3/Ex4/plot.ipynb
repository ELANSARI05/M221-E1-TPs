```python
import matplotlib.pyplot as plt

# Data
threads = [1, 2, 4, 8, 16]
cpu_times = [0.047008, 0.037259, 0.086735, 0.043462, 0.042478]

# Plot
plt.figure(figsize=(8,5))
plt.plot(threads, cpu_times, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Threads')
plt.ylabel('CPU Time (seconds)')
plt.title('Matrix Multiplication CPU Time vs Number of Threads')
plt.xticks(threads)
plt.grid(True)
plt.show()
```
