# Exercise 4
## Objective
Evaluate the performance improvement of the parallel MPI implementation compared to the serial version.

---

## Results

- **Serial Time:** 0.034924 seconds  
- **Parallel Time (3 processes):** 0.014142 seconds  
- **Max Error:** 0.000000e+00  

The zero max error confirms that the parallel implementation is correct.

---

## Speedup

Speedup formula:

Speedup = T_serial / T_parallel

Speedup = 0.034924 / 0.014142 ≈ 2.47

---

## Efficiency

Efficiency formula:

Efficiency = Speedup / Number_of_Processes

Efficiency = 2.47 / 3 ≈ 0.82

Efficiency ≈ 82%

---

## Interpretation

- Parallel execution significantly reduces runtime.
- Speedup (2.47) is close to the ideal value (3).
- Efficiency (82%) shows good utilization of computational resources.
- Small performance loss is due to communication and synchronization overhead.

---

## Conclusion

The MPI parallel implementation improves performance while maintaining correctness.  
The program demonstrates good scalability and efficient use of 3 processes.
